## 1.1 Immutable types

Immutable types are basically those types whose value can't be changed after its creation.

Immutable types include int, float, str, bool, tuple, None.

When you are assigning new integer to the same variable, you aren't changing the value but essentially assinging same variable to a different int object with a new value. The old integer value still exsists in memory but doesn't have any variable that points to it.

Since you can't change immutable types, you need to actually declare new object itself if you want your variable to represent a new value.

````python
# We have a sample integer here
apple_price = 20

# Let's check the memory id of the object that this variable points to
print(id(apple_price))

# Let's change the integer value of the variable
apple_price = 40

# Let's observe the memory id again
print(id(apple_price))
````

## 1.2 Mutable types

Mutable types on the other hand can be freely changed after it is created.

These includes collections like list, dict, set.

If you change any elements of these kinds of data types, since they are mutable, no new objects are created.

<div align="center">
  <br>
  <img src="https://github.com/aksho-sh/CollabImages/blob/main/VariableMutable.png?raw=true" style="margin: 20px;">
  <p style="margin-top: 10px;"><em>Figure 1: Mutable types</em></p>
</div>

````python
# We have a sample list here
fruit_list = ["apple", "mango", "orange"]

# Let's check the memory id of the object that this variable points to
print(id(fruit_list))

# Let's change the value of the list by appending a new value to the list
fruit_list.append("guava")

# Let's check the memory id of hte object that this variable points to now
print(id(fruit_list))

# Let's now assign a new list to the variable
fruit_list = ["peach", "strawberry"]

# Let's check the memory id of the object that this variable points to
print(id(fruit_list))
````

### 1.3 Copying mutable types

````python
fruit_list = ["apple", "mango", "orange"]
fruit_list2 = fruit_list.copy()
print(fruit_list)
print(fruit_list2)
fruit_list.append('grape')

print(fruit_list)
print(fruit_list2)
````

````python
# If assinging one mutable type to another variable just creates a reference, then how do you actually copy it?

# You can use .copy() method on the variable

fruit_list = ["apple", "mango", "orange"]

print(fruit_list)
print(f"fruit_list points to {id(fruit_list)}")

fruit_list_copy = fruit_list.copy()

# You can see that it now points to a different memory address
print(fruit_list_copy)
print(f"fruit_list_copy points to {id(fruit_list_copy)}")

fruit_list.append("grapes")

print(fruit_list)
print(fruit_list_copy)
````

## 1.3 Assigning one variable to another variable

When you assign one variable to another, and change the second variable, the behaviour depends based on wether the object type that the variable refers to is mutable or immutable.


<div align="center">
  <br>
  <img src="https://github.com/aksho-sh/CollabImages/blob/main/ImmutableVsMutable.png?raw=true" style="margin: 20px;">
  <p style="margin-top: 10px;"><em>Figure 1: Assigning one variable to another</em></p>
</div>

````python
# Immutable objects behaviour

integer_1 = 20
integer_2 = 40

integer_1 = integer_2

# both of them are 20 now and point to the same object

print(f"Memory id of integer_1 is {id(integer_1)}")
print(f"Memory id of integer_2 is {id(integer_2)}")

integer_2 = 50

# integer_1 remains 20 despite change in integer_2
print(integer_1)
print(integer_2)

# integer_2 now points to new memory address and integer_1 to the same
print(f"Memory id of integer_1 is {id(integer_1)}")
print(f"Memory id of integer_2 is {id(integer_2)}")
````

````python
# Mutable objects behaviour

fruit_list_1 = ["mango", "apple", "blueberry"]
fruit_list_2 = [None]

fruit_list_2 = fruit_list_1

# both of them are now pointing to the same object
print(f"Memory id of integer_1 is {id(fruit_list_1)}")
print(f"Memory id of integer_2 is {id(fruit_list_2)}")

fruit_list_2.append("orange")

print(fruit_list_1)
print(fruit_list_2)

# both of them are still poitning to the same object
print(f"Memory id of integer_1 is {id(fruit_list_1)}")
print(f"Memory id of integer_2 is {id(fruit_list_2)}")
````

## 2. Control Flow

Control flow dictates the sequential flow of the program. Python offers similar suite of control flow to other languages.


Before we move to control flow, we should remember that in case of python, it uses indentation instead of curly braces to seperate a block of code.

A code like this in c++:

```cpp
if (x>0){
  int y = 10;
  printf("%d\n", y);
}
```

Would look like this in python:

```python
if x>0:
  y = 10;
  print(y)
```

Here, the indentation defines what falls inside the context of the if block. As soon as the indentation returns to the base level, the context of if statement ends.

## 2.1 If Statement

````python
# If statement is similar to other languages with option of else if and else, just difference is the elif instead of else if as syntax. And use of colon (:)

age = 10

if age < 20 and age >= 13:
  print("Person is  a teen")
elif age <13:
  print("Person is a child")
else:
  print("Person is an adult")
````

## 2.2 Boolean operators and comparison operators

````python
# Boolean operator includes and, or and not
# This is different to other languages where &&, || and ! symbols are used.

# Comparison operators inlcude <, >, <=, >=, == (equals to), != (not equals)

integer_value = 20

if integer_value < 30 and integer_value > 10:
  print("The value is  between 20 and 30")

if integer_value % 2 == 0 or integer_value % 3 == 0:
  print("The value is divisible by 2 or 3")

if integer_value % 3 != 0:
  print("The value isn't divisble by 3")

is_present = True

if not is_present:
  print("Yes")
````

## 2.3 Truthiness

Empty collection objects like lists, dictionaries, along with 0 and None are Falsy. This means that when used with statements like If, they resolve to false.

This can be useful to check if the list is empty or the value returned by a function is 0 or None.

````python
# Checking falsiness

# Empty list falsiness
a = [1,2,3]

# a resolves to True here since a isn't empty.
if a:
  print("The list is full")

a = []

if a:
  print("This list is empty")

# a resolves to False here since a is full.
# if not a:
#   print("The list is empty")

# other similar cases

a = None

if a:
  print("A is None")

a = 0

if a:
  print("A is 0")
````

## 2.4 For loops

For loops similar to other languages exist in Python. They take two forms, both of which we can observe below.

````python
# One way to work with For loop is to directly loop through a collection

fruit_list = ["apple", "orange", "mango"]

# a string is extracted from list each turn and stored in fruit variable
for fruit in fruit_list:
  print(fruit)
````

````python
# Another way to work with For loop is with range() function

for index in range(0, 10, 2):
  print(index)

# This is similar to for syntax of c++ with for (x=0, x<10, x=x+2). This range provides initial variable, the terminating condition and increment amount as
# as parameters.

# If your for loop just needs index values starting from 0 to a certain point at increament of 1, you can shorten the syntax

print("This is the new loop")

for index in range(8):
  print(index)
````

## 2.5 While loop, breaks and continue

Python also offers while loop similar to other languages, which loops based on condition.

It also has constructs like breaks and continue which can be used with both while and for.

````python
# A simple while loop in python

a = 0

# Loops until a is less than 15
while a<15:
  a += 1

  # if a is divisible by 0 skips the loop from this point, nothing below here gets executed if the condition is true


  print(a)

  # if a equals 11, the loop breaks and stops executing
  if a == 11:
    break
````

## 2.6 Enumerate, and Zip

Enumerate and Zip fulfills distinct purpose in loops.

If you need an index value while looping through a list, you use Enumerate. If you need to enumerate 2 equal list at once, and access values from both the list sequentially, you use zip.

````python
# Enumerate example

fruit_list = ["apple", "orange", "mango"]

for index, fruit in enumerate(fruit_list):
  print(f"{fruit} is in {index} position")
````

````python
# Zip example

fruit_list = ["apple", "orange", "mango"]
fruit_price = [10, 20, 15]

for fruit, price in zip(fruit_list, fruit_price):
  print(f"{fruit} costs Rs.{price}")
````

### 3.1 Default Arguments and Keyword Arguments

Python allows you to define default arguments which come with default values. These are handy if you want there to be a default value for a parameter incase the user doesn't pass the value.

````python
# Function with a default parameter
def calculate_cost(quantity, price_of_apple = 20):
  return quantity * price_of_apple

# You have the option of either passing the price or not, uses default value if not passed
print(f"The cost of 10 apples  are {calculate_cost(10)}")
````

````python
# Passing values with keyword

def add_3_numbers(number1, number2, number3, caller):
  sum = number1 + number2 + number3
  print(f"function called by {caller}")
  return sum

# You can either pass the values in order while calling the function
add_3_numbers(10, 20, 40, "Sam")

print("\ncalled with keyword arguments")
# Or you can pass the values in different order, but define the order
add_3_numbers(number1=10, caller="Sam", number3=40, number2=10)

# You can also mix this approach, but keyword arguments must always follow unnamed arguments
add_3_numbers(10, 20, caller="Sam", number3=40)

# It can't be, this will give an error, you can uncomment and try it out
# add_3_numbers(10, 20, caller="Sam", 40)
````

### 3.2 Mutable datatype as function parameter

Passing a mutable datatype as a function argument follows the same logic as assigning mutable datatype from one variable to another we discussed earlier. It takes the reference, hence modifying the mutable datatype inside the function modifies it outside the function too.

````python
def add_score(scores):
  scores.append(10)
  print(f"This is printend from inside the function: {scores}")
````

````python
score_list = [10,20,30,40,50]

add_score(score_list)

print(f"This is printed from out side the fuction {score_list}")
````

#### 5.4 Sum of even numbers

Write a function `sum_evens(numbers)` that takes a list of numbers and returns the sum of only the even ones.

Test it with `[1, 2, 3, 4, 5, 6, 7, 8]`.

````python
# Your function here
def sum_evens(numbers):
    total = 0

    for num in numbers:
        if num % 2 == 0:
            total += num

    return total

# Test
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
print(sum_evens(numbers))

````

#### 5.8 Fruit inventory

Given the fruit list and prices below, write a function `expensive_fruits(fruits, prices, threshold)` that returns a list of fruit names whose price is strictly greater than `threshold`.

For example, with `threshold=15`, the result should be `["mango", "guava"]`.

*Hint: you'll need `zip` and an empty list you append to.*

````python
fruit_list = ["apple", "mango", "orange", "guava"]
fruit_prices = [10, 30, 15, 25]

# Your function here
def expensive_fruits(fruits, prices, threshold):
    result = []

    for fruit, price in zip(fruits, prices):
        if price > threshold:
            result.append(fruit)

    return result

# Test
print(expensive_fruits(fruit_list, fruit_prices, 15))
````

#### 5.9 Average score with the math module

Using `import math`, write a function `average_rounded(numbers)` that returns the average of a list of numbers, **rounded down** to the nearest integer using `math.floor()`.

For example, the average of `[10, 20, 25]` is `18.33...`, and `math.floor(18.33)` is `18`.

````python
# Your code here
import math

def average_rounded(numbers):
    average = sum(numbers) / len(numbers)
    return math.floor(average)

# Test
print(average_rounded([10, 20, 25]))
````

#### 5.10 Stretch

Given a list of student scores, write a function `score_summary(scores)` that returns a dictionary with the keys `"count"`, `"highest"`, `"lowest"`, and `"average"`.

For example, with `[78, 92, 65, 88, 91]` the result might look like:
```python
{"count": 5, "highest": 92, "lowest": 65, "average": 82.8}
```

*Hint: `min()`, `max()`, `sum()`, and `len()` are built-ins that work on lists.*

````python
# Your code here
def score_summary(scores):
    summary = {
        "count": len(scores),
        "highest": max(scores),
        "lowest": min(scores),
        "average": sum(scores) / len(scores)
    }
    return summary

# Test
scores = [78, 92, 65, 88, 91]

print(score_summary(scores))
````
