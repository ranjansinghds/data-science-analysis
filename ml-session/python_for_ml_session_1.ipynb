## 1. Python Execution model

### How machine runs your code

Let's start with the fundamental concept, your computer doesn't run the code you give it directly. First of all, it converts the code into some sort of binary form (represented in form of 0s and 1s). There are 2 approaches that programming languages takes for this.

### Compilers

Compiler based languages take your source code and covert it machine understandable binary file. These binary file can directly be executed in your computer. The problem with this is that a binary file that is compiled on one computer isn't guranteed to run on another computer since computer architectures vary from one to another.

Languages like **C++ and C** falls under this category.

<div align="center">
  <br>
  <img src="https://github.com/aksho-sh/CollabImages/blob/main/Compiler.png?raw=true" style="margin: 20px;">
  <p style="margin-top: 10px;"><em>Figure 1: Compiler based execution</em></p>
</div>

### Interpreters

Interpreter based languages on the other hand avoid direct compilation to machine code. Many of them take the code that you have written, compile them into a byte code that the interpreter understands. This interpreter then takes over the responsibility of executing this code and the user doesn't have to worry about machine architectures. This makes interpreter based languages much more portable as any machine with the interpreter can execute this code.

<div align="center">
  <br>
  <img src="https://github.com/aksho-sh/CollabImages/blob/main/Interpreter.png?raw=true" style="margin: 20px;">
  <p style="margin-top: 10px;"><em>Figure 1: Interpreter based execution</em></p>
</div>

**Python and JavaScripts** are an example of interpreted language.

Every browser has an engine which parses, interprets the code and  executes the code.

Python scripts that you write is converted into bytecode by its interpreter CPython. CPython then walks through the bytecode and executes the bytecode one by one until it has nothing left to execute.

### 2.1.2 Order of cell execution

The order in which the cells are executed matters tremendously.

If you declare in cell 1:

```python
x = 20
```

and in cell 2:

```python
x = 30
```

in cell 3:

```python
print(x)
```

The value of x printed will be based on which cell is run last before the print statement.

Running cell 1 -> cell 2 -> cell 3, will give 30 as output.

Running cell 2 -> cell 1 -> cell 3, will give 20 as output.

You can see the order of the cells executed, to the left of the cell in between the [ ].

Try this out yourself.

````python
# Step 1: define x as 5. Run this cell.
x = 5
````

````python
# Step 2: define x again as 10. Run this cell.
x = 10
````

````python
# Step 3: Print the value of x
# 1. Run Step 1 first, run step 2 second and run step 3 third
# 2. Run Step 2 first, run step 1 second and run step 3 third

print(x)

# You should see different values of x based on the execution order of 1 and execution order of 2. What happens is the last executed cell overwrites the values
# of the previous cell when the variable names are the same.
````

```cpp
// C++ code
```


In python, variables don't have type. You can assign any type of values to a variable.

### 3.2.1 Arithmetic

Most arithmetic works exactly like you'd expect.

````python
#Addition

number_1 = 6
number_2 = 3

# Addition
sum = number_1 + number_2
print("The sum is: ", sum)

# Substraction
difference = number_1 - number_2
print("The difference is: ", difference)

# Product
product = number_1 * number_2
print("The product is: ", product)

# Quotient
quotient = number_1 / number_2
print("The quotient is: ", quotient)

# Remainder
remainder = number_1 % number_2
print("The remainder is: ", remainder)

# Exponent ^
exponent = number_1 ** number_2
print("The exponent is: ", exponent)
````

Divsion between / and // operators.

One gives actual division value in floating point. This returns value as float.

Another gives a floored value. This returns value as integer.

````python
# Regular division always gives a float — even when the result is a whole number
print(10 / 2)     # 5.0 — note the .0, this is a float
print(7 / 2)      # 3.5 - gives a more precise value

# Integer division (//) drops the decimal part 3.5 -> 3, doesn't give 4 by rouding up
print(10 // 2)    # 5 — an int
print(7 // 2)     # 3 — drops the .5
````

The / division gives a precise value. With // there is a loss of information. However, this is desirable sometimes.

````python
# Calculating total number of boxes that can be fully packed where capacity of each box is 10 items

total_items = 47
total_no_of_full_boxes = 47//10
print(total_no_of_full_boxes)
````

## 4. Strings

Strings are text. You can use single or double quotes — they mean the same thing. Triple quotes let you write multi-line strings.

````python
single = 'hello'
double = "hello hello"
multi = """This string
spans multiple lines."""

print(single)
print(double)
print(multi)
````

### 4.1 f-strings

When you want to put values inside a string, use an **f-string**. Put `f` before the opening quote, and write `{expression}` anywhere you want a value substituted in.

This is the modern convention for the purpose of printing any expression or values in the print function.

````python
name = "Bishal"
score = 87.456

# Instead of
print("Student: ",name, " Score: ", score)

# Use
print(f"Student: {name}, Score: {score}")

# You can format numbers — :.2f means "float with 2 decimal places"
print(f"Score: {score:.2f}")

# You can compute things inside the braces
print(f"Score doubled: {score * 2}")
print(f"Passed: {score >= 50}")
print(f"Passed: {"yes" if score>=40 else "fail"}")

#condition ? statement1 : statement2
````

### 4.2 Useful string methods

Strings have methods you call with the dot syntax: `string.method()`. Here are the ones that actually come up in data work.

````python
text = "  Hello, World!  "
print(text.strip())       # "Hello, World!" — removes whitespace from both ends
print(text.lower())       # converts to lowercase
print(text.upper())       # converts to uppercase
print(text.replace("World", "Python"))  # find/replace


# split() turns a string into a list, breaking at the given separator
csv_line = "alice,bob,charlie,dave"
names = csv_line.split(",")
print(names)

# join() goes the other way — list of strings into one string
joined = " | ".join(names)
print(joined)
````

### 4.3 String indexing and slicing

Strings are sequences of characters, and you can access them by position. You will see this same syntax used for lists and numpy arrays.

- Indexing starts at `0` (the first character is at index `0`).
- Negative indices count from the end: `-1` is the last character.
- **Slicing** with `[start:end]` gives you a chunk. `end` is *exclusive*.

<div align="center">
  <br>
  <img src="https://github.com/aksho-sh/CollabImages/blob/main/PythonStrings.png?raw=true" style="margin: 20px;">
  <p style="margin-top: 10px;"><em>Figure 1: Python Strings</em></p>
</div>

````python
s = "machine learning"

print(s[0])        # 'm' — first character
print(s[7])        # ' ' — the space
print(s[-1])       # 'g' — last character
print(s[-7])       # 'l' — 7th from the end

print(s[0:7])      # 'machine' — chars 0 through 6 (7 is excluded)
print(s[7:len(s)])

# Step: every Nth character
print(s[::2])      # 'mcielann' — every 2nd char
print(s[::-1])     # 'gninrael enihcam' — reversed!
````

## 5. Lists

A **list** is an ordered collection of values. Create one with square brackets. Lists can contain anything, including a mix of types, including other lists.

````python
scores = [78, 92, 65, 88, 91, 73]
names = ["Alice", "Bob", "Charlie"]
mixed = [1, "hello", 3.14, True, None]      # legal, but usually not a good idea
empty = []

print(scores)
print(names)
print(mixed)
print(empty)

# How many things in the list?
print(len(scores))

# len() function can also be used with string to show the number of letters in the string
````

### 5.1 Indexing and slicing

Indexing and slicing in lists works in same way to indexing and slicing in string.

````python
scores = [78, 92, 65, 88, 91, 73]

print(scores[0])       # 78
print(scores[-1])      # 73 — last element
print(scores[2:5])     # [65, 88, 91]
print(scores[:3])      # [78, 92, 65] — first three
print(scores[-2:])     # [91, 73] — last two
````

### 5.2 Modifying lists

Unlike strings, lists are **mutable**. That means you can change them after creating them.

````python
scores = [78, 92, 65, 88, 65]

# Change an element by index
scores[0] = 100
print(scores)

# Add an element to the end of the list
scores.append(95)
print(scores)

# Add a given list to the list
scores.extend([80, 85])
print(scores)

# Remove an element by value
scores.remove(65)
print(scores)

# Sort in place
scores.sort()
print(scores)
````

### 5.3 List operations

````python
scores = [78, 92, 65, 88, 91, 73]

# Is something in the list?
print(78 in scores)       # True
print(100 in scores)      # False

# Quick stats — built-in functions that work on any iterable
print(len(scores))        # how many
print(min(scores))        # smallest
print(max(scores))        # largest
print(sum(scores))        # total
print(sum(scores) / len(scores))  # average
````

### 5.4 Nested List

Lists can contain lists. This is how you'd represent tabular data in pure Python.

````python
# A grid — each inner list is a row
grid = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(grid[0])           # [1, 2, 3] — first row
print(grid[0][0])        # 1 — first row, first column
print(grid[2][1])        # 8 — third row, second column
````

## 6. Dictionaries

A **dictionary** (or `dict`) maps **keys** to **values**. If a list is "a sequence of things indexed by position," a dict is "a sequence of things indexed by name."

Create one with curly braces and `key: value` pairs.

<div align="center">
  <br>
  <img src="https://github.com/aksho-sh/CollabImages/blob/main/PythonDictionary.png?raw=true" style="margin: 20px;">
  <p style="margin-top: 10px;"><em>Figure 1: Python Dictionaries</em></p>
</div>

````python
student = {
    "name": "Alice",
    "age": 20,
    "scores": [85, 90, 78],
    "passed": True
}

print(student)

# Access by key, not by position
print(student["name"])
print(student["scores"])

# Add or change a key
student["grade"] = "A"
student["age"] = 21
print(student)

# Check if a key exists
print("name" in student)
print("email" in student)
````

You will see dicts everywhere in ML work:

- A row of a pandas DataFrame behaves like a dict.
- JSON data from APIs comes as dicts.
- Model configurations are usually dicts.
- Metrics are usually returned as dicts.

Getting comfortable with `dict_name["key"]` access is essential.

### 6.1 Getting keys, values, and items

````python
student = {"name": "Alice", "age": 20, "grade": "A"}

print(list(student.keys()))     # ['name', 'age', 'grade']
print(list(student.values()))   # ['Alice', 20, 'A']
print(list(student.items()))    # [('name', 'Alice'), ('age', 20), ('grade', 'A')]
````

### 6.2 Safe access with `.get()`

If you access a key that doesn't exist with `[]`, Python raises a `KeyError`. Use `.get()` when you're not sure if the key exists, it returns `None` (or a default you specify) instead of crashing.

````python
student = {"name": "Alice", "age": 20}

# This would raise KeyError if uncommented:
#print(student["email"])

# This is safe — returns None if the key doesn't exist
print(student.get("email"))

# You can also provide a default value
print(student.get("email", "no email on file"))
````

## 7. Tuples

A **tuple** is an ordered collection like a list, but with one big difference: tuples are **immutable**. Once created, you can't change them — no `append`, no item assignment.

Create them with parentheses (or sometimes no brackets at all).

````python
point = (3, 5)             # a tuple
rgb = (255, 100, 50)
single = (42,)             # one-element tuple needs the trailing comma

print(point[0])            # 3 — indexing works like a list
print(rgb[-1])             # 50

# This would error — tuples can't be modified:
#point[0] = 99
````

### 7.1 Tuple unpacking
When a function returns multiple values, it actually returns a tuple, and you can "unpack" it directly into multiple variables. This is the same multiple-assignment pattern you saw earlier.

````python
# Unpacking a tuple
point = (3, 5)
x, y = point
print(x, y)

x = point[0]
y = point[1]

# This is why you'll see code like this in ML:
# X_train, X_test, y_train, y_test = train_test_split(...)
# The function returns a tuple of four things and you unpack them by name.

# Another common pattern — shape of an array:
shape = (150, 4)    # 150 rows, 4 columns
rows, cols = shape
print(f"Rows: {rows}, Cols: {cols}")
````

### When to use which container

Rough rules of thumb you'll refine over time:

- **List** : When you want to group a similar type of values in an order.
- **Dict** : Default choice when each item has a unique name.
- **Tuple** : A fixed list-like object, that won't change. Default choice for small fixed groupings (a point, a shape, a return value of a function).
