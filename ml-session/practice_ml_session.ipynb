### Splitting and Joining strings

Strings can be broken down into list of strings, based on delimiter. For example: you can break down a string with a sentence into a list of word.

````python
#The syntax to break down string using split
sentence = "This is the sample sentence we want to be broken down"
list_of_words = sentence.split(" ") # This represents a space
print(list_of_words)
````

The list of string can be combined back using by invoking **.join()** on string you want to join with.

````python
combined_sentence = " ".join(list_of_words) #Joins by space " "
print(combined_sentence)
````

````python
# Task 6
# Create a string with the first names seperated by comma: "name1,name2,name3,name4"
names_string = "Ram,Shyam,John,Smith"

# Split the string by invoking .split() method using appropriate delimiter value that is seperating the strings
list_of_names = names_string.split(",") # This represents a comma
print(list_of_names)

# Combine back the strings by using - value on the list of strings you obtained using the .split() method
combined_names = "-".join(list_of_names) # Joins by hyphen "-"
print(combined_names)
````

### List slicing

List slicing is a critical operation in python to obtain a sub list from the given list by specifying the starting and ending index of the list.

```python
list_1 = ["1", "10", "3", "4", "5","6", "7", "8", "9"]

list_2 = list_1[2:6] # This creates a new list with values starting from index 2 upto index 5 (excluding index 6)

print(list_2)

list_3 = list_1[:6] # This creates a new list from index 0 to index 5 (excluding index 6)
print(list_3)

list_4 = list_1[2:] # This creates list from index 2 to the last item of the list
print(list_4)
```

````python
# Task 8
# You are given following list

fruit_list = ["apple", "mango", "grape", "guava", "orange", "strawberry", "dragon fruit", "avacado"]

# Create another list using slicing operation from mango to dragon fruit using slicing operation (include dragon fruit)
fruit_list_1 = fruit_list[1:7] # This represents from index 1 to index 6 (include index 6)
print(fruit_list_1)

# Create another list using slicing operation from apple to orange using slicing operation (include orange)
fruit_list_2 = fruit_list[0:5] # This represents from index 0 to index 4 (include index 4)
print(fruit_list_2)

# create another list using slicing operation from orange to avacado using slicing operation (include avacado)
fruit_list_3 = fruit_list[4:] # This represents from index 4 to index 7 (include index 7)
print(fruit_list_3)
````

## Tuple

Tuples are just like list, but you can't modify them after creation. It is usually created to return values multiple from a function in uneditable format.

Tuples can be created using:

```python
tuple_1 = (1,2,3)
```

To use the tuple value, you must unpack the tuple value to the variables. That can be done using:

```python
x,y,z = tuple_1 #tuple value from respective tuple index will be assigned to respective variables based on the order of each

# x - takes first element
# y - takes second element and so on
```

````python
# Task 10

# Create a tuple of 4 integer values
tuple_of_integers = (10, 20, 30, 40)

# Unpack the tuple values to 4 different variables
a, b, c, d = tuple_of_integers

# Calculate the average value of the variables
average = (a + b + c + d) / 4
print(average)
````
