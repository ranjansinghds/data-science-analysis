# What is Data Analysis?
Data analysis is a process of inspecting, cleaning, transforming and modeling data with the goal of discovering useful information, informing conclusions and suppoting decision-making.

## Data Analysis Process
* step-1: Problem Understanding/ Business Understanding
* Step-2: Data Wrangling/ Data Munging
* Step-3: Exploratory Data Analysis(EDA)
* Step-4: Feature Engineering
* Step-5: Drawing Conclusions
* Step-6: Communicating Results/ Data Storytelling

# 1. Problem Understanding / Business Understanding
Before touching data, understand:

* What problem are we solving?
* What is the goal?
* What type of insights are needed?
* What decisions will be made from the analysis?<br>

Example:
* Why are sales decreasing?
* Which customers are likely to churn?
* Which product performs best?

# 2. Data Wrangling/Data Munging

Raw data is usually:

* messy
* incomplete
* inconsistent
* duplicated
* unstructured<br>

Data Wrangling is the process of converting that messy data into clean and useful data.

### 1. Data Gathering

Data Gathering is the process of collecting data from different sources for analysis.

##### Sources of Data
CSV files, Excel files, Databases, APIs, Websites, Surveys, Sensors<br>

Example:

Collecting customer sales data from:<br>
company database, online forms, website logs

### 2. Data Assessing

Data Assessing is the process of inspecting and evaluating data to identify quality issues, structure, and usefulness.

##### It helps us understand:

* Is the data complete?
* Are there missing values?
* Are data types correct?
* Are there duplicates?
* Is the data usable?

##### Common Checks
* Missing values
* Incorrect data types
* Duplicate records
* Inconsistent formatting
* Outliers

### 3. Data Cleaning

Data Cleaning is the process of fixing or removing incorrect, incomplete, duplicate, or inconsistent data.

##### Tasks
* Handling missing values
* Removing duplicates
* Fixing data types
* Correcting text formats
* Handling outliers

# 3.Exploratory Data Analysis (EDA):
Exploratory Data Analysis (EDA) is the process of examining, summarizing, and visualizing data to understand its patterns, relationships, trends, and problems before performing further analysis or building machine learning models.

## Main Purpose of EDA
* Understand dataset structure
* Find missing values
* Detect anomalies/outliers
* Analyze distributions
* Discover correlations
* Generate insights

## Types of Exploratory Data Analysis (EDA)
EDA (Exploratory Data Analysis) is mainly divided into:

* Univariate Analysis
* Bivariate Analysis
* Multivariate Analysis

### 1. Univariate Analysis
Analysis of a single variable/column at a time.

#### Purpose:

* Understand distribution
* Find patterns
* Detect outliers
* Check spread of data

##### Example
Analyzing only:
`Age`

#### Numerical Data Methods
* Mean
* Median
* Mode
* Standard deviation
##### Common Plots
Histogram, Box Plot, Density Plot

#### Categorical Data Methods
* Frequency count
* Percentage distribution
##### Common Plots
Bar Chart, Pie Chart
##### Example:
Using Seaborn:
`sns.histplot(df['Age'])`

### 2. Bivariate Analysis
Analysis of relationship between two variables.

#### Purpose:

* Find correlation
* Compare variables
* Identify trends/relationships

##### Examples
`Age vs Salary`
`Gender vs Purchase`

### Types
Numerical vs Numerical<br>
Plots
Scatter Plot, Line Plot
#### Categorical vs Numerical
Plots
* Bar Plot
* Box Plot
* Violin Plot

#### Categorical vs Categorical
Plots
* Count Plot
* Heatmap (cross-tab)
##### Example
`sns.scatterplot(x='Age', y='Salary', data=df)`

### 3. Multivariate Analysis
Analysis of more than two variables together.

#### Purpose:

* Understand complex relationships
* Detect interactions
* Find hidden patterns

#### Example
`Age + Salary + Experience + Department`

### Common Techniques
Pair Plot, Heatmap, 3D Scatter Plot, Correlation Matrix
##### Example
`sns.pairplot(df)`

# 4. Feature Engineering
Feature Engineering is the process of creating, transforming, or selecting variables (features) to improve machine learning model performance.

### Goals
* Improve accuracy
* Create meaningful features
* Convert raw data into useful input

## Common Feature Engineering Tasks
* Encoding categorical data
* Scaling
* Normalization
* Creating new columns
* Date-time feature extraction
* Binning
* Log transformation

# 5.Drawing Conclusions
Drawing Conclusion means making final interpretations, decisions, or insights from analyzed data.

### It is the process of answering:

* “What did we learn from the data?”
* “What does the analysis tell us?”

## Purpose of Drawing Conclusions
* Convert analysis into meaningful insights
* Support business decisions
* Solve problems
* Recommend actions

# 6.Communicating Results / Data Storytelling
Communicating Results or Data Storytelling means presenting data insights, findings, and conclusions in a clear, understandable, and meaningful way using visuals, explanations, and narratives.

#### It helps people understand:

* what the data says
* why it matters
* what actions should be taken

##### Simple Meaning
After analyzing data, you explain the results to:

* managers
* clients
* teams
* stakeholders

`Data Storytelling = Data + Visualization + Explanation(Narrative)`
