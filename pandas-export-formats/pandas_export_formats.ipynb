# Pandas Export
* to csv
* to excel
* to html
* to json
* to sql

```python
import pandas as pd
df = pd.DataFrame()
```

```python
df = pd.read_csv('deliveries.csv')
df.head()
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df.to_csv('batsman_runs.csv', index=False)
```

```python
df.pivot_table(index='batsman', columns='bowling_team', values='batsman_runs', aggfunc='sum').to_csv('batsman_runs_by_team.csv')
```

# to_excel

```python
df
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
temp_df
```

```python
temp_df.to_excel('batsman_runs.xlsx')
```

```python
temp_df.to_excel('output.xlsx', sheet_name='Batsman Runs', index=False)
```

```python
temp_df2 =df.pivot_table(index='batsman', columns='bowling_team', values='batsman_runs', aggfunc='sum')
```

```python
with pd.ExcelWriter('output.xlsx') as writer:
    temp_df.to_excel(writer, sheet_name='Batsman Runs', index=False)
    temp_df2.to_excel(writer, sheet_name='Runs by Team')
```

# to_html

```python
df
```

```python
df.query('batsman_runs == 6').pivot_table(index='over', columns='ball', values='batsman_runs', aggfunc='sum').to_html('sixes_heatmap.html')
```

# to_json

```python
df
```

```python
df.groupby(['batting_team','batsman'])['batsman_runs'].sum().unstack().to_json('ipl.json')
```
