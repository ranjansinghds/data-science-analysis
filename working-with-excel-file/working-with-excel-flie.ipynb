```python
# duplicate values analysis
df.duplicated().sum()
```

## 5. Unique Values

```python
df['Product Category'].unique()
```

```python
df['Region'].nunique()
```

## 6. Value Counts

```python
# Most Frequent Categories
df['Product Category'].value_counts()
```

```python
# Region Frequency
df['Region'].value_counts()
```

## 7. Sorting

```python
# sorting values
df.sort_values('Sales', ascending=False)
```

## 8. Filtering

```python
# Technology Category
df[df['Product Category'] == 'Technology']
```

```python
# High Profit Orders
df[df['Profit'] > 500]
```

## 9. GroupBy (MOST IMPORTANT)

```python
# Total Sales by Category
df.groupby('Product Category')['Sales'].sum()
```

```python
# Average Profit by Region
df.groupby('Region')['Profit'].mean()
```

```python
# multiple aggregations
df.groupby('Product Category')['Sales'].agg(['sum', 'mean', 'max', 'min'])
```

## 10. correlation Analysis

```python
df.corr(numeric_only=True)
```

## 11. Top N Analysis

```python
# Top 5 Sales
df.nlargest(5, 'Sales')
```

```python
# Lowest Profit
df.nsmallest(5, 'Profit')
```

## 12. Pivot Table (VERY IMPORTANT)

```python
pd.pivot_table(df, values='Sales', index='Product Category',columns='Region',aggfunc='sum')
```

## 13. Date Analysis

```python
# If dataset contains order dates:
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Extract Year from Order Date
df['Year'] = df['Order Date'].dt.year
```

```python
# Monthly Sales
df.groupby(df['Order Date'].dt.month)['Sales'].sum()
```

## 14. Visualization Functions

```python
# Bar Plot
df.groupby('Product Category')['Sales'].sum().plot(kind='bar')
```

```python
# Histogram
df['Sales'].plot(kind='hist')
```

```python
# Scatter Plot
import matplotlib.pyplot as plt
plt.scatter(df['Sales'], df['Profit'])
```
